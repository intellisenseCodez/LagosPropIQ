```python
import requests
```

```python
# let's make our html output readable by importing pprint
from pprint import pprint
```

```python
base_res_url = requests.get("https://jiji.ng/lagos/houses-apartments-for-sale")
base_res_url
```

```python
base_response = base_res_url.content
```

```python
# now let's convert the response to Beautiful soup in a more readable format

from bs4 import BeautifulSoup
```

```python
# now let's create an object from the BeautifulSoup as we know it's a coding convention that all classes starts with a capital letter

html_soup = BeautifulSoup(base_response, "html.parser")
# html_soup
```

```python
product_div_container = html_soup.find("div", class_="masonry-wall b-list-advert__gallery")
len(product_div_container)
```

```python
all_product_mansory = product_div_container.find_all("div", class_="masonry-item")
print(len(all_product_mansory))
print("==" * 40)
# pprint(all_product_mansory)
```

```python
# let's use a loop to get all PRODUCT ITEMS FOR THE BASE URL
all_items_per_page = []

for product_items in all_product_mansory:
    title = product_items.find("div", class_="b-list-advert-base__data__title").text
    prop_price = product_items.find("div", class_="b-list-advert-base__data__price").text
    prop_description = product_items.find("div", class_="b-list-advert-base__description").text
    prop_location = product_items.find("span", class_="b-list-advert__region__text").text
    prop_size_sqm = product_items.find("div", class_="b-list-advert-base__item-attr").text
    

    all_product_mansory_items = {"property_title": title, "price": prop_price, 
                          "description": prop_description, "location": prop_location, "size_sqm": prop_size_sqm}
    all_items_per_page.append(all_product_mansory_items)
```
