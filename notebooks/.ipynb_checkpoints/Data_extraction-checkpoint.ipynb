```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re
import os
from geopy.geocoders import Nominatim

# ============================================
# CONFIG
# ============================================
SAVE_PATH = r"C:\Users\THIS-PC\Documents\LagosPropIQ\Data\raw\landsng_lagos_properties.csv"
BASE_URL = "https://land.ng/search-results/?keyword=&states%5B%5D=lagos&page="
HEADERS = {"User-Agent": "Mozilla/5.0"}
geolocator = Nominatim(user_agent="lagos_land_scraper")

# Make sure the folder exists
os.makedirs(os.path.dirname(SAVE_PATH), exist_ok=True)

# Storage
all_data = []

# ============================================
# SCRAPE UNTIL NO MORE PAGES
# ============================================
page = 1

while True:
    print(f"\nğŸŒ Scraping page {page} ...")
    url = BASE_URL + str(page)

    # Retry mechanism
    for attempt in range(3):
        try:
            response = requests.get(url, headers=HEADERS, timeout=40)
            if response.status_code == 200:
                break
        except Exception as e:
            print(f"âš ï¸ Attempt {attempt+1}/3 failed for page {page}: {e}")
            time.sleep(4)

    else:
        print(f"âŒ Failed to load page {page} after 3 retries. Stopping.")
        break

    soup = BeautifulSoup(response.text, "html.parser")
    listings = soup.find_all("div", class_="item-wrap")

    if not listings:
        print("âœ… No more listings found. Scraping completed.")
        break

    for item in listings:
        try:
            # Title + Link
            title_tag = item.find("h2", class_="item-title")
            title = title_tag.get_text(strip=True) if title_tag else None
            link = title_tag.find("a")["href"] if title_tag and title_tag.find("a") else None

            # Price
            price_tag = item.find("span", class_="price")
            price = price_tag.get_text(strip=True) if price_tag else None
            price_num = float(re.sub(r"[^\d.]", "", price)) if price else None

            # Location
            location_tag = item.find("address", class_="item-address")
            location = location_tag.get_text(strip=True) if location_tag else None

            # Land Size + Type
            land_size = None
            land_type = None

            amenity_tags = item.find_all("li", class_="item-amenities-item")
            for li in amenity_tags:
                text = li.get_text(strip=True).lower()

                if "sqm" in text or "square" in text:
                    land_size = li.get_text(strip=True)
                elif "plot" in text or "estate" in text or "land" in text:
                    land_type = li.get_text(strip=True)

            # Clean numeric sqm
            land_size_num = None
            if land_size:
                clean_size = re.sub(r"[^\d.]", "", land_size)
                land_size_num = float(clean_size) if clean_size else None

            # Average â‚¦/sqm
            avg_price_sqm = None
            if price_num and land_size_num:
                avg_price_sqm = round(price_num / land_size_num, 2)

            # Coordinates
            lat, lon = None, None
            if location:
                try:
                    geo = geolocator.geocode(f"{location}, Lagos, Nigeria", timeout=20)
                    if geo:
                        lat, lon = geo.latitude, geo.longitude
                except:
                    pass

            # Append row
            all_data.append({
                "Title": title,
                "Link": link,
                "Price": price,
                "Price (â‚¦)": price_num,
                "Location": location,
                "Land Type": land_type,
                "Land Size": land_size,
                "Land Size (sqm)": land_size_num,
                "Average â‚¦/sqm": avg_price_sqm,
                "Latitude": lat,
                "Longitude": lon
            })

        except Exception as e:
            print(f"âš ï¸ Error parsing listing: {e}")

    # Be polite
    time.sleep(5)
    page += 1

# ============================================
# SAVE MERGED RESULTS
# ============================================
df_final = pd.DataFrame(all_data)
df_final.to_csv(SAVE_PATH, index=False, encoding="utf-8-sig")

print("\nğŸ‰ SCRAPING COMPLETE!")
print(f"ğŸ“Œ Total listings gathered: {len(df_final)}")
print(f"ğŸ“ Saved to: {SAVE_PATH}")
```
